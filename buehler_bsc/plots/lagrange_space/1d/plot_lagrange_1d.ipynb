```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1D Local basis functions

```python
# load the local basis data
local_basis_data = pd.read_csv('~1D_lagrange_local_basis.csv', sep=',') 
```

```python
plt.clf()

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(12, 4))

axs = [ax0, ax1]
for i in range(0, 2):
    axs[i].plot(local_basis_data['x'], local_basis_data[f"v_{i}"], label=f"$\\varphi_{i}$")
    axs[i].set_title(f"1D Lagrange (1st order) local shape function $\\varphi_{i}$")

plt.show()
```
