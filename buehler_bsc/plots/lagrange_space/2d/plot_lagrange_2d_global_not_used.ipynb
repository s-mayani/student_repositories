In the end this plot was not used, and the functions that generated this data were removed as well from the implementation.
For completeness sake, I chose to keep it in the repository.

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata
from matplotlib.patches import Polygon
```

```python
# Load the data
df = pd.read_csv('~2D_lagrange_global_basis.csv')
x = df['x'].values
y = df['y'].values
z = df.iloc[:, 2].values
```

```python
# Define grid
xi = np.linspace(-1, 1, 1000)
yi = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(xi, yi)

# Interpolate surface
Z = griddata((x, y), z, (X, Y), method='cubic')

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Show plot
plt.show()
```
