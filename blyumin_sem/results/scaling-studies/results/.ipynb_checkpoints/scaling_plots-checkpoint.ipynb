```python
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter, NullFormatter, FixedLocator
from typing import Dict, List
import matplotlib.ticker as mticker
    
def _setup_subplot_axes(ax, x_label, y_label, title, x_col_data, log_y=False):
    """
    Helper function to set up common axes properties for subplots,
    including gridlines only on major axes.
    """
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
#     ax.set_title(title)

    # --- Gridline Customization ---
    ax.grid(True, which="major", ls="--", c='0.7') # Show only major gridlines with dashed style
    ax.grid(False, which="minor") # Explicitly hide minor gridlines

    # --- X-axis Customization (always set to log for scaling plots, with literal ticks) ---
    ax.set_xscale('log')
    unique_x_values = sorted(x_col_data.dropna().unique())
    # Filter for integer values to ensure ticks are clean, assuming ranks/GPUs are integers
    valid_x_ticks = [x for x in unique_x_values if pd.notna(x) and x == int(x)]
    if valid_x_ticks:
        ax.set_xticks(valid_x_ticks)
        ax.xaxis.set_major_formatter(ScalarFormatter()) # Show actual numbers (2, 4, 8)
        ax.xaxis.set_minor_formatter(NullFormatter()) # No minor labels for intermediate values
        ax.xaxis.set_major_locator(FixedLocator(valid_x_ticks)) # Ensure these specific ticks are shown
    else:
        # Fallback if no valid integer ticks found
        ax.get_xaxis().set_major_formatter(mticker.ScalarFormatter())
        ax.ticklabel_format(axis='x', style='plain') # Prevent scientific notation on x-axis if possible


    # --- Y-axis Customization ---
    if log_y:
        ax.set_yscale('log')
        ax.yaxis.set_major_formatter(ScalarFormatter()) # Show actual numbers (10, 100)
        ax.yaxis.set_minor_formatter(NullFormatter()) # No minor labels for intermediate values
    else:
        ax.set_yscale('linear')
        
    ax.tick_params(axis='x', labelsize=12, width=1.5, length=6) # Adjust labelsize, width, and length as needed
    ax.tick_params(axis='y', labelsize=12, width=1.5, length=6) # Adjust labelsize, width, and length as needed
    ax.tick_params(axis='x', which='minor', width=0.8, length=3) # Adjust width and length as needed
    ax.tick_params(axis='y', which='minor', width=0.8, length=3) # Adjust width and length as needed

        
def set_global_plot_text_sizes(base_font_size=12):
    """
    Sets global matplotlib rcParams to control text sizes across all plots.

    Args:
        base_font_size (int): The base font size. Other font sizes will be
                              derived from this to maintain proportionality.
    """
    # General font size for text elements
    plt.rcParams['font.size'] = base_font_size
    plt.rcParams['axes.labelsize'] = base_font_size * 1.1 # Slightly larger than base    
    plt.rcParams['axes.titlesize'] = base_font_size * 1.2 # Larger than labels
    plt.rcParams['xtick.labelsize'] = base_font_size * 1.1 # Slightly smaller than base
    plt.rcParams['ytick.labelsize'] = base_font_size * 1.1 # Slightly smaller than base
    plt.rcParams['legend.fontsize'] = base_font_size * 1.1 # Consistent with tick labels
    plt.rcParams['figure.titlesize'] = base_font_size * 1.3 # Largest for main figure title

    print(f"Matplotlib global text sizes set with base font size: {base_font_size}")

set_global_plot_text_sizes(base_font_size=16)
```

```python
BASE_DIR = os.getcwd()

def parse_timing_file(filepath: str) -> Dict[str, float]:
    '''
    Extracts key numbers from a timing_*.dat file.
    
    Returns a df containing the structured data.
    '''
    data: Dict[str, float] = {}

    try:
        with open(filepath, "r", encoding="utf-8") as f:
            lines: List[str] = f.readlines()

        in_section = False
        for line in lines:
            if "================================================" in line:
                in_section = True
                continue
            if in_section and line.strip() and not line.startswith("="):
                parts = line.split()
                if len(parts) >= 5:
                    try:
                        wall_avg = float(parts[-2])
                        wall_min = float(parts[-3])
                        wall_max = float(parts[-4])
                    except ValueError:
                        continue
                    # Process name = everything before the last 4 columns
                    name = " ".join(parts[:-4]).replace("...............", "").strip()
                    data[f"{name} Wall max"] = wall_max
                    data[f"{name} Wall min"] = wall_min
                    data[f"{name} Wall avg"] = wall_avg
    except FileNotFoundError:
        print(f"[parse_timing_file] File not found: {filepath}")
    except Exception as exc:
        print(f"[parse_timing_file] Error parsing {filepath}: {exc}")

    return data

rex_cpu  = r"^(?P<n_particles>\d+)_particles_timing_(?P<n_ranks>\d+)(?:\.dat)?$"
rex_gpu  = r"^(?P<n_particles>\d+)_particles_(?P<n_nodes>\d+)_timing_\d+(?:\.dat)?$"

# 
def process_directory(dir_name: str, regex: str, extract_fn):
    '''
    Iterates through all the files in the folder and 
    adds results to a data frame
    
    Returns:
    Dataframe containing all the concatenated results in a directory 
    
    '''
    full_dir = os.path.join(BASE_DIR, dir_name)
    df = pd.DataFrame()
    if not os.path.isdir(full_dir):
        print(f"[process_directory] {full_dir} missing â€“ skipped.")
        return df

    pattern = re.compile(regex)
    for filename in sorted(os.listdir(full_dir)):
        m = pattern.match(filename)
        if not m:
            print("no match")
            print(filename)
            continue
        row = extract_fn(m)

        # ensure n_nodes exists for CPU cases
        if "n_nodes" not in row:
            row["n_nodes"] = row["n_ranks"]

        timing = parse_timing_file(os.path.join(full_dir, filename))
        if timing:
            row.update(timing)
            df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
    return df

extract_cpu = lambda m: {
    "n_particles": int(m.group("n_particles")),
    "n_ranks":    int(m.group("n_ranks")),
}

extract_gpu = lambda m: {
    "n_particles": int(m.group("n_particles")),
    "n_nodes":     int(m.group("n_nodes")),
    "n_ranks":     int(m.group("n_nodes")),   # equals n_nodes
}

print("Processing flattened result directoriesâ€¦")

# Strong (CPU) ------------------------------------------------------
df_strong_1 = process_directory(dir_name="results_strong_1",
    regex=rex_cpu,
    extract_fn=extract_cpu,
)

df_strong_2 = process_directory(dir_name="results_strong_2",
    regex=rex_cpu,
    extract_fn=extract_cpu,
)

df_strong_3 = process_directory(dir_name="results_strong_3",
    regex=rex_cpu,
    extract_fn=extract_cpu,
)

# Strong (GPU) ------------------------------------------------------
df_strong_gpu_1 = process_directory(
    dir_name="results_strong_gpu_1",
    regex=rex_gpu,
    extract_fn=extract_gpu,
)

df_strong_gpu_2 = process_directory(
    dir_name="results_strong_gpu_2",
    regex=rex_gpu,
    extract_fn=extract_gpu,
)

df_strong_gpu_3 = process_directory(
    dir_name="results_strong_gpu_3",
    regex=rex_gpu,
    extract_fn=extract_gpu,
)

# Weak (CPU) --------------------------------------------------------
df_weak_1 = process_directory(
    dir_name="results_weak_1",
    regex=rex_cpu,
    extract_fn=extract_cpu,
)

df_weak_2 = process_directory(
    dir_name="results_weak_2",
    regex=rex_cpu,
    extract_fn=extract_cpu,
)

df_weak_3 = process_directory(
    dir_name="results_weak_3",
    regex=rex_cpu,
    extract_fn=extract_cpu,
)

# df_weak_fix = process_directory(
#     dir_name="results_weak",
#     regex=rex_cpu,
#     extract_fn=extract_cpu,
# )

# Weak (GPU) --------------------------------------------------------
df_weak_gpu_1 = process_directory(
    dir_name="results_weak_gpu_1",
    regex=rex_gpu,
    extract_fn=extract_gpu,
)

df_weak_gpu_2 = process_directory(
    dir_name="results_weak_gpu_2",
    regex=rex_gpu,
    extract_fn=extract_gpu,
)

df_weak_gpu_3 = process_directory(
    dir_name="results_weak_gpu_3",
    regex=rex_gpu,
    extract_fn=extract_gpu,
)

# df_weak_gpu_fix = process_directory(
#     dir_name="results_weak_gpu",
#     regex=rex_cpu,
#     extract_fn=extract_cpu,
# )

TIMER_COL = "Hermiticity Timer... Wall avg"

def summarise_runs(run_dfs, timer_col=TIMER_COL):
    '''
    Concatenate repeated runs and return a tidy DataFrame with the mean
    and standard deviation of the chosen timer.
    '''
    df = pd.concat(run_dfs, ignore_index=True)

    summary = (
        df.groupby(["n_particles", "n_nodes", "n_ranks"], as_index=False)[timer_col]
          .agg(["mean", "std"])
          .reset_index()
          .rename(columns={"mean": "hermiticityTimer_avg",
                           "std":  "hermiticityTimer_std"})
    )
    
    return summary

# ---------------------------------------------------------------
#  Build the four final DataFrames
# ---------------------------------------------------------------
df_strong       = summarise_runs([df_strong_1,       df_strong_2,       df_strong_3])
df_strong_gpu   = summarise_runs([df_strong_gpu_1,   df_strong_gpu_2,   df_strong_gpu_3])
df_weak         = summarise_runs([df_weak_1,         df_weak_2,         df_weak_3])
df_weak_gpu     = summarise_runs([df_weak_gpu_1,     df_weak_gpu_2,     df_weak_gpu_3])
# df_weak_fix     = summarise_runs([df_weak_fix])
# df_weak_gpu_fix = summarise_runs([df_weak_gpu_fix])
# ------------------------------------------------------------------
#  Optional: persist to CSV
# ------------------------------------------------------------------
# df_strong.to_csv("df_strong.csv", index=False)
# df_strong_gpu.to_csv("df_strong_gpu.csv", index=False)
# df_weak.to_csv("df_weak.csv", index=False)
# df_weak_gpu.to_csv("df_weak_gpu.csv", index=False)

print("\n--- df_strong (first 5 rows) ---")
print(df_strong.head())
print("\n--- df_strong_gpu (first 5 rows) ---")
print(df_strong_gpu)
print("\n--- df_weak (first 5 rows) ---")
print(df_weak.head())
print("\n--- df_weak_gpu (first 5 rows) ---")
print(df_weak_gpu.head())

```

```python
def plot_combined_strong_scaling(df: pd.DataFrame,
                                 x_col: str,
                                 y_col_time: str,
                                 yerr_col_time: str,
                                 title: str,
                                 x_label: str,
                                 y_label: str,
                                 group_by_col: str = 'n_particles',
                                 save_path: str = None,
                                 log_scale_y: bool = False):
    '''
    Generates three separate figures for strong scaling, showing Wall Time, Speedup, and Efficiency.
    '''

    group_values = sorted(df[group_by_col].unique())
    colors = plt.cm.viridis(np.linspace(0, 1, len(group_values)))

    # Store max ranks observed for ideal lines
    max_ranks_observed = 0

    # Data structures to hold plot elements for each figure's legend
    handles_time, labels_time = [], []
    handles_speedup, labels_speedup = [], []
    handles_efficiency, labels_efficiency = [], []

    # Create figures and axes for each plot type
    fig_time, ax_time = plt.subplots(figsize=(10, 7))
    fig_speedup, ax_speedup = plt.subplots(figsize=(10, 7))
    fig_efficiency, ax_efficiency = plt.subplots(figsize=(10, 7))

    for i, group_val in enumerate(group_values):
        subset_df = df[df[group_by_col] == group_val].sort_values(by=x_col)

        # --- Prepare data for plotting ---
        # Get T(1) for speedup/efficiency calculation
        t1_row = subset_df[subset_df[x_col] == 1]

        T1, sigma1 = t1_row[y_col_time].iloc[0], t1_row[yerr_col_time].iloc[0]

        # Calculate Speedup and its error
        S = T1 / subset_df[y_col_time]
        sigmaTp = subset_df[yerr_col_time]
        sigmaS = S * np.sqrt((sigma1 / T1)**2 + (sigmaTp / subset_df[y_col_time])**2)

        # Calculate Efficiency and its error
        p = subset_df[x_col]
        E = T1 / (p * subset_df[y_col_time])
        sigmaE = E * np.sqrt((sigma1 / T1)**2 + (sigmaTp / subset_df[y_col_time])**2)

        # --- Plotting ---

        # Plot 1: Wall Time
        line_time = ax_time.errorbar(subset_df[x_col], subset_df[y_col_time], yerr=subset_df[yerr_col_time],
                                      marker='o', markersize='10', linestyle='-', color=colors[i],
                                      label=f'{group_by_col.replace("n_", "# ").replace("_", " ").title()}: {group_val}')
        handles_time.append(line_time)
        labels_time.append(f'{group_by_col.replace("n_particles", "Nx ").replace("_", " ").title()}: {group_val}')

        # Plot 2: Speedup
        line_speedup = ax_speedup.errorbar(subset_df[x_col], S, yerr=sigmaS,
                                             marker='o', markersize='10', linestyle='-', color=colors[i])
        handles_speedup.append(line_speedup)
        labels_speedup.append(f'{group_by_col.replace("n_particles", "Nx ").replace("_", " ").title()}: {group_val}')

        # Plot 3: Efficiency
        line_efficiency = ax_efficiency.errorbar(subset_df[x_col], E, yerr=sigmaE,
                                                   marker='o', markersize='10', linestyle='-', color=colors[i])
        handles_efficiency.append(line_efficiency)
        labels_efficiency.append(f'{group_by_col.replace("n_particles", "Nx ").replace("_", " ").title()}: {group_val}')

        # Keep track of largest number of processors observed
        max_ranks_observed = max(max_ranks_observed, subset_df[x_col].max())

    # --- Add Ideal Lines ---
    if max_ranks_observed > 0:
        x_ideal = np.logspace(0, np.log10(max_ranks_observed * 1.1), 100)

        # Ideal speedup
        ax_ideal_speedup_line, = ax_speedup.plot(x_ideal, x_ideal, 'k--', alpha=0.7, label='Ideal Speedup')
        handles_speedup.append(ax_ideal_speedup_line)
        labels_speedup.append('Ideal')

        # Ideal efficiency
        ax_ideal_efficiency_line, = ax_efficiency.plot(x_ideal, np.ones_like(x_ideal), 'k--', alpha=0.7, label='Ideal Efficiency')
        handles_efficiency.append(ax_ideal_efficiency_line)
        labels_efficiency.append('Ideal')

    # --- Setup subplot titles and labels for each figure ---
    _setup_subplot_axes(ax_time, x_label, "Wall Time (s)", f"{title} - Wall Time", df[x_col], log_y=log_scale_y)
    _setup_subplot_axes(ax_speedup, x_label, "Speedup", f"{title} - Speedup", df[x_col], log_y=True) # Speedup usually log y
    _setup_subplot_axes(ax_efficiency, x_label, "Efficiency", f"{title} - Efficiency", df[x_col], log_y=False) # Efficiency usually linear y

    # --- Add legends to each figure ---
    ax_time.legend(handles=handles_time, labels=labels_time, title='Problem Size', loc='best', frameon=True)
    ax_speedup.legend(handles=handles_speedup, labels=labels_speedup, title='Problem Size / Ideal Lines', loc='best', frameon=True)
    ax_efficiency.legend(handles=handles_efficiency, labels=labels_efficiency, title='Problem Size / Ideal Lines', loc='best', frameon=True)

    # Adjust layout for each figure
    fig_time.tight_layout()
    fig_speedup.tight_layout()
    fig_efficiency.tight_layout()

    # Save each figure
    if save_path:
        base, ext = os.path.splitext(save_path)
        save_path_time = f"{base}_wall_time{ext}"
        save_path_speedup = f"{base}_speedup{ext}"
        save_path_efficiency = f"{base}_efficiency{ext}"

        try:
            fig_time.savefig(save_path_time, bbox_inches='tight')
            print(f"Wall Time plot saved successfully to {save_path_time}")
        except Exception as e:
            print(f"Error saving Wall Time plot to {save_path_time}: {e}")

        try:
            fig_speedup.savefig(save_path_speedup, bbox_inches='tight')
            print(f"Speedup plot saved successfully to {save_path_speedup}")
        except Exception as e:
            print(f"Error saving Speedup plot to {save_path_speedup}: {e}")

        try:
            fig_efficiency.savefig(save_path_efficiency, bbox_inches='tight')
            print(f"Efficiency plot saved successfully to {save_path_efficiency}")
        except Exception as e:
            print(f"Error saving Efficiency plot to {save_path_efficiency}: {e}")

    plt.show()

# -- Weak scaling: TIME / RANK -----------------------------------------------
def plot_combined_weak_scaling(df: pd.DataFrame,
                               x_col: str,
                               y_col: str,
                               yerr_col: str,
                               title: str,
                               x_label: str,
                               save_path: str = None,
                               group_by_col: str = 'n_particles',
                               log_scale_y: bool = False):
    '''
    Generates three separate figures for weak scaling: Total Wall Time, Time Per Rank, and Combined Ideal Scaling Reference.
    '''

    group_values = sorted(df[group_by_col].unique())
    colors = plt.cm.viridis(np.linspace(0, 1, len(group_values)))

    all_p_values = sorted(df[x_col].dropna().unique())

    # Calculate ideal weak scaling reference values
    min_p_value = df[x_col].min()
    base_time_per_rank = df[df[x_col] == min_p_value][y_col].mean() / min_p_value
    ideal_x_values = np.array(all_p_values)

    # --- Figure 1: Total Wall Time ---
    fig1, ax1 = plt.subplots(figsize=(7, 6)) # Create a new figure and axis for Total Wall Time
    handles1, labels1 = [], [] # Initialize lists for legend handles and labels for this figure

    # Sort the entire DataFrame by x_col to plot a single connected line
    df_sorted = df.sort_values(by=x_col)

    # Prepare data for plotting for the single connected line
    p_all = df_sorted[x_col]
    T_total_all = df_sorted[y_col]
    sigma_T_total_all = df_sorted[yerr_col]
#     print(T_total_all)
    # Plot 1: Total Wall Time (single connected line)
    ax_total_time_line = ax1.errorbar(p_all, T_total_all, yerr=sigma_T_total_all,
                                    marker='o', markersize='10', linestyle='-', color='blue',
                                    label='Measured Scaling')
    handles1.append(ax_total_time_line[0])
    labels1.append('Measured Scaling')


    # Ideal Weak Scaling line for Total Wall Time (should be constant)
    ideal_weak_scaling = ax1.plot(ideal_x_values, [base_time_per_rank] * len(ideal_x_values),
                   'k--', label='Ideal Weak Scaling')
    handles1.append(ideal_weak_scaling[0])
    labels1.append('Ideal Weak Scaling')

#     if 'GPU' in title:
#         ax1.set_ylim(5e-5,8e-5) # Apply original code's specific y-axis limit if 'GPU' is in title

    # Setup axis properties for Figure 1
    _setup_subplot_axes(ax1, x_label, "Total Wall Time (s)", "Total Wall Time", df[x_col], log_y=log_scale_y)
    # Add legend to Figure 1
    ax1.legend(handles=handles1, labels=labels1, title='Weak Scaling Series', loc='best', frameon=True)
    fig1.tight_layout() # Adjust layout for Figure 1

    if save_path:
        try:
            # Modify save_path for the first figure
            save_path_total = save_path.replace('.png', '_total_wall_time.png')
            plt.savefig(save_path_total, bbox_inches='tight')
            print(f"Total Wall Time plot saved successfully to {save_path_total}")
        except Exception as e:
            print(f"Error saving Total Wall Time plot to {save_path_total}: {e}")
       
    plt.show() # Display Figure 1 and continue execution

    # --- Figure 2: Time Per Rank (Now Efficiency) ---
    fig2, ax2 = plt.subplots(figsize=(7, 6)) # Create a new figure and axis for Time Per Rank
    handles2, labels2 = [], [] # Initialize lists for legend handles and labels for this figure

    T1_for_efficiency = df[df[x_col] == min_p_value][y_col].iloc[0]

    # Sort the entire DataFrame by x_col to plot a single connected line
    df_sorted = df.sort_values(by=x_col)

    # Prepare data for plotting for the single connected line
    p_all = df_sorted[x_col]
    T_total_all = df_sorted[y_col]
    sigma_T_total_all = df_sorted[yerr_col]

    efficiency_all = T1_for_efficiency / T_total_all

    # Error propagation for efficiency: sigma(E) = E * sqrt((sigma(T1)/T1)^2 + (sigma(T_total)/T_total)^2)
    # Assuming T1_sigma is the error for T_total at min_p_value
    T1_sigma_val = df_sorted[df_sorted[x_col] == min_p_value][yerr_col].iloc[0]
    sigma_efficiency_all = efficiency_all * np.sqrt((T1_sigma_val/T1_for_efficiency)**2 + (sigma_T_total_all/T_total_all)**2)

    # Plot 2: Efficiency (single connected line)
    ax_efficiency_line = ax2.errorbar(p_all, efficiency_all, yerr=sigma_efficiency_all,
                                    marker='o', markersize='10', linestyle='-', color='blue',
                                    label='Measured Scaling')
    handles2.append(ax_efficiency_line[0])
    labels2.append('Measured Scaling')

    # Add Ideal Scaling line to Efficiency plot (should be a constant line at 1)
    ideal_efficiency_line = ax2.plot(ideal_x_values, [1] * len(ideal_x_values), 'k--', label='Ideal Scaling')
    handles2.append(ideal_efficiency_line[0])
    labels2.append('Ideal Scaling')

    # Setup axis properties for Figure 2
    _setup_subplot_axes(ax2, x_label, "Efficiency", "Efficiency", df[x_col], log_y=log_scale_y)
    # Add legend to Figure 2
    ax2.legend(handles=handles2, labels=labels2, title='Weak Scaling Series', loc='best', frameon=True)
    fig2.tight_layout() # Adjust layout for Figure 2

    if save_path:
        try:
            # Modify save_path for the second figure
            save_path_per_rank = save_path.replace('.png', '_efficiency.png')
            plt.savefig(save_path_per_rank, bbox_inches='tight')
            print(f"Weak Scaling Efficiency plot saved successfully to {save_path_per_rank}")
        except Exception as e:
            print(f"Error saving Weak Scaling Efficiency plot to {save_path_per_rank}: {e}")
    plt.show()
```

```python
x_col = 'n_ranks'
y_col = 'hermiticityTimer_avg'
yerr = 'hermiticityTimer_std'

#results for df_strong
plot_combined_strong_scaling(
        df_strong,
        x_col, # This should be the column with MPI ranks
        y_col,
        yerr,
        title='Strong Scaling CPU: Hermiticity Timer Average Time v\s. Number of MPI Ranks',
        x_label='No. MPI Ranks',
        y_label='Hermiticity Timer Average Time (s)',
        group_by_col='n_particles', # Plot separate lines for each particle count
        save_path='strong_scaling_cpu_hermiticity.png',
        log_scale_y=True
    )

plot_combined_weak_scaling(
    df_weak,
    x_col, # This should be the column with GPU nodes
    y_col,
    yerr,
    title='',
    x_label='No. MPI Ranks',
    group_by_col='n_particles', # Plot separate lines for each particle count (if multiple series)
    save_path='weak_scaling_cpu_hermiticity', # Adjusted filename
    log_scale_y=False
)
```
